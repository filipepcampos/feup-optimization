```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("allocations.csv", encoding="latin-1")
df["Day"] = df["Day"].map(lambda x: int(x[4:]))
df.head()
```

```python
print("[")
for day in df["Day"].unique():
    unique_candidate_ids = df[df["Day"] == day]["Candidate Id"].unique()
    percentage_distribution = np.random.normal(4, 5, 5000)

    count = 0
    for missing_percentage in percentage_distribution:
        missing_percentage = 0 if missing_percentage < 0 else missing_percentage
        missing_staff_count = int(missing_percentage/100 * len(unique_candidate_ids))
        
        if missing_staff_count <= 0 or missing_staff_count > 20:
            continue
        count += 1

        missing_staffs = list(np.random.choice(unique_candidate_ids, missing_staff_count, replace=False))
        print("{", f"{','.join(str(i) for i in missing_staffs)}", "},")

        if count == 30:
            break
print("]")
```
